# Converting the Raw data to a dataframe

```python
t=['SN_T{:01d}'.format(i)for i in range(1,10)]
ra=['SN_RA{:01d}'.format(i)for i in range(1,10)]
la=['SN_LA{:01d}'.format(i)for i in range(1,10)]
rl=['SN_RL{:01d}'.format(i)for i in range(1,10)]
ll=['SN_LL{:01d}'.format(i)for i in range(1,10)]
s=t+ra+la+rl+ll
c=[]
for h in range(1,126):
  for i in s:
    c.append(i+'_{0:1d}'.format(h))
df=pd.DataFrame(columns= c + ['activity'])
# Specify the path to the data
data_path = 'data'

# Iterate through the activities
for activity in range(1, 20):
    activity_path = os.path.join(data_path, 'a{:02d}'.format(activity))
    # Iterate through the subjects
    for subject in range(1, 9):
        subject_path = os.path.join(activity_path, 'p{}'.format(subject))
        # Iterate through the segments
        for segment in range(1, 61):
            segment_path = os.path.join(subject_path, 's{:02d}.txt'.format(segment))
            # Read the sensor data for the segment
            data= np.loadtxt(segment_path, delimiter=',').flatten()
            df=df.append(pd.DataFrame([data],columns=c))
            df.iloc[len(df)-1,-1]=activity
df=df.reset_index(drop=True)
```
